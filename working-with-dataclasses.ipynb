```python
import sqlite3
con = sqlite3.connect("person.db")
```

```python
cur = con.cursor()
```

```python
query = "CREATE TABLE person(id, name, age, job, phone, address)"
```

```python
res = cur.execute(query)
```

```python
for idx, data in enumerate(database):
    cur.execute("INSERT INTO person VALUES(?, ?, ?, ?, ?, ?)", (idx,) + astuple(data))

```

```python
con.commit()  # Remember to commit the transaction after executing INSERT.
```

```python
result = cur.execute("SELECT * FROM person").fetchall()
```
